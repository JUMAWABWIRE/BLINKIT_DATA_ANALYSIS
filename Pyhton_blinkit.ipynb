```python
#Cleaning.
print(df['Item Fat Content'].unique())
```

```python
df['Item Fat Content'].replace({'LF': 'Low Fat',
                                'low fat': 'Low Fat',
                                'reg': 'Regular'}, inplace= True)
```

```python
print(df['Item Fat Content'].unique())
```

### KPIS REQUIREMENTS.


```python
#SALES
total_sales = df['Sales'].sum()

#Average sales.

avg_sales = df['Sales'].mean()

#Total number of items sold.
no_of_items_sold = df['Sales'].count()

#Ratings Average.
avg_ratings = df['Rating'].mean()

print(f"Total Sales: ${total_sales:,.1f}")
print(f"Aaverage Sales: ${avg_sales:,.0f}")
print(f"Average Rating: {avg_ratings:,.1f}")
print(f"Number of items Sold: {no_of_items_sold:,.0f}")
```

### CHARTS REQUIREMENT.
## 1. Total Sales by Fat Content.

```python
Sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(Sales_by_fat,
        labels = Sales_by_fat.index,
        autopct= '%.0f%%',
        startangle= 90
                      )

plt.title('Sales by fat Content')
plt.axis('equal')
plt.show()
```

# 2. Total Sales Per Item.

```python
Sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize= (12, 8))
bars= plt.bar(Sales_by_type.index, Sales_by_type.values)

plt.xticks(rotation = 90)
plt.xlabel= 'Item Type'
plt.ylabel ='Total Sales'
plt.title = 'Total Sales By Item Type'
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
    f'{bar.get_height():,.0f}', ha= 'center', va='bottom',fontsize = 8)
plt.tight_layout()
plt.show()
    
```

# 3. Fat Contebt by Outlet.

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular', 'Low Fat']]

x_axis = grouped.plot(kind='bar', figsize=(10,7), title= 'Fat content by Outlet')
plt.xlabel = 'Outlet Location'
plt.ylabel = 'Total Sales'
plt.legend(title= 'Item Fat Content')
plt.tight_layout()
plt.show()
```

# 4. Sales per Outlet estableshment.

```python
Sales_per_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(Sales_per_year.index,Sales_per_year.values,marker='o',linestyle='-'
)

plt.xlabel = "Outlet Establishment Year"
plt.ylabel = 'Total Sales'
plt.title = 'Total Outlet Establishment Year'

# Loop through points to annotate
for x, y in zip(Sales_per_year.index, Sales_per_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

# 4. Sales by outlet size.

```python
Sales_by_Size = df.groupby('Outlet Size')['Sales'].sum()

plt.pie(Sales_by_Size,
        labels = Sales_by_Size.index,
        autopct= '%1.2f%%',
        startangle= 90
                      )

plt.title = 'Sales By Outlet Size'
plt.figure(figsize=(4,4))
plt.show()
```

# 5. Sales by Outlet Locations.

```python
Sales_Location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
Sales_Location = Sales_Location.sort_values('Sales', ascending = True )
ax = sns.barplot(x ='Sales', y='Outlet Location Type', data=Sales_Location)
plt.title('Sales by Outlet Location Type.')
plt.xlabel = 'Totals Sales'
plt.ylabel = "Outlet Location Type"
plt.figure(figsize=(10, 4))
plt.tight_layout()
plt.show()
```
